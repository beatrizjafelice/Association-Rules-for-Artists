# Atividade

Extraindo os dados:

```python
import pandas as pd

df = pd.read_csv('https://drive.google.com/u/3/uc?id=1aDEAX-LqUtdLl10n5SrjqF-dMAmDBmTE&export=download', names=['bandas'])
df.head()
```

Em análises de regras de associação, as linhas do conjunto de dados são chamadas de "transações". Cada transação é composta por um ou mais itens. Nesse caso, as transações representam os artistas favoritos dos alunos e cada item corresponde a um artista.

Separando os itens em listas:

```python
data = list(df['bandas'].apply(lambda x: x.split(',')))
data
```

Removendo espaços extras para evitar duplicidades:

```python
for lista in data:
    for i, nome in enumerate(lista):
        lista[i] = nome.strip()
data
```

Agora, codifica-se o conjunto de dados como uma tabela atributo-valor. Nessa tabela, cada linha representa uma transação e cada coluna representa um item. Os valores True ou False indicam se o item da coluna está presente naquele registro ou não.

```python
from mlxtend.preprocessing import TransactionEncoder

t_enc = TransactionEncoder()
t_enc_data = t_enc.fit(data).transform(data)
t_enc_data
```

Transformando em um dataframe para melhor visualização:

```python
df_encoded = pd.DataFrame(t_enc_data, columns=t_enc.columns_)
df_encoded
```

Em seguida, extrai-se os *itemsets*, que são os conjuntos de itens que aparecem com mais frequência na base. Os *itemsets* podem ser formados por um ou mais itens. O parâmetro *min_support* determina a frequência mínima do itemset na base para ser considerado frequente.

```python
from mlxtend.frequent_patterns import apriori, association_rules

# OBS: em bases de dados grandes, aumentar o min_support
# (mínima proporção de transações em que o itemset aparece)
freq_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)
freq_itemsets
```

Extraindo as regras de associação:

```python
regras = association_rules(freq_itemsets, metric='confidence', min_threshold=0.6)
regras
```

A tabela de regras de associação apresenta relações entre artistas e bandas, mostrando com que frequência eles são ouvidos juntos. Aqui está uma interpretação detalhada de cada coluna na tabela:

1. **Antecedents (Antecedente)**: O artista ou banda que aparece em primeiro lugar na regra.
2. **Consequents (Consequente)**: O artista ou banda que aparece em segundo lugar na regra.
3. **Antecedent support**: A proporção de transações que incluem o antecedente.
4. **Consequent support**: A proporção de transações que incluem o consequente.
5. **Support (Suporte)**: A proporção de transações que incluem tanto o antecedente quanto o consequente.
6. **Confidence (Confiança)**: A probabilidade de o consequente ser ouvido quando o antecedente é ouvido.
7. **Lift**: A razão entre a confiança da regra e o suporte do consequente. Compara o suporte observado e o suporte esperado se o antecedente e o consequente fossem independentes. Um *lift* maior que 1 indica uma associação positiva.
8. **Leverage**: A diferença entre o suporte observado e o suporte esperado se o antecedente e o consequente fossem independentes. Valores próximos de 1 indicam associação positiva, próximos de 0 indicam nenhuma associação e próximos de -1, associação negativa.
9. **Conviction**: Relação entre a frequência esperada do antecedente sem o consequente e a frequência observada do antecedente sem o consequente. Indica quanto mais frequentemente o antecedente implica o consequente em comparação com os casos onde o antecedente não implica o consequente. Valores altos indicam associação positiva.
10. **Métrica de Zhang**: Analisa como a presença e a não presença do antecedente impactam a presença do consequente. Pode assumir valores entre -1 e 1, assim como *leverage*.
